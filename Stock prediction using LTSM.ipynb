```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
data = pd.read_csv("AAPL.csv")
```

```python
train_data = data[:int(0.8*len(data))]
test_data = data[int(0.8*len(data)):]
```

```python
x_train = train_data["Close"].to_numpy().reshape(-1, 1)
y_train = train_data["Open"].to_numpy()
```

```python
x_test = test_data["Close"].to_numpy().reshape(-1, 1)
y_test = test_data["Open"].to_numpy()
```

```python
model = Sequential()
model.add(LSTM(50, activation="tanh", input_shape=(1, 1)))
model.add(Dense(1))
```

```python
model.compile(loss="mse", optimizer="adam")
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
predictions = model.predict(x_test)
```

```python
accuracy = np.mean(np.abs(predictions - y_test))
print("Accuracy:", accuracy)
```
